```python
# we used same code for multiple cities for more trends in the data set refer to pptx file
import pandas as pd
import numpy as np
df=pd.read_csv("bombay-modified-year.csv")
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['month'], df['maxtempC'],color='g',linestyle='solid');
plt.xticks(rotation = 25)
plt.xlabel('Month',color='w')
plt.ylabel('maxtempC',color='w')
plt.title('Weather Report(Bombay)', fontsize = 20, color='w')
plt.tick_params(axis='x', colors='white')    
plt.tick_params(axis='y', colors='white')  
plt.grid()
plt.show()
```

```python
plt.scatter(df['month'], df['mintempC'],color='g',linestyle='solid');
plt.xticks(rotation = 25)
plt.xlabel('Month',color='w')
plt.ylabel('mintempC',color='w')
plt.title('Weather Report(Bombay)', fontsize = 20, color='w')
plt.tick_params(axis='x', colors='white')    
plt.tick_params(axis='y', colors='white')  
plt.grid()
plt.show()
```
